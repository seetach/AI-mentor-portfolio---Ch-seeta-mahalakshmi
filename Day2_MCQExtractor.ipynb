```python
import json
import re

# Read the text file
with open("10_mcq_questions_with_answers.txt", "r", encoding="utf-8") as file:
    content = file.read()

# Split questions
question_blocks = re.split(r"\n(?=\d+\.)", content.strip())

mcqs = []

for block in question_blocks:
    lines = block.strip().split("\n")

    # Question
    question_match = re.match(r"(\d+)\.\s*(.*)", lines[0])

    if question_match:
        q_no = int(question_match.group(1))
        question_text = question_match.group(2)

        options = {}
        answer = ""

        # Options and answer
        for line in lines[1:]:
            option_match = re.match(r"([A-D])\.\s*(.*)", line)

            if option_match:
                options[option_match.group(1)] = option_match.group(2)

            elif line.startswith("Answer:"):
                answer = line.split(":")[1].strip()

        mcq = {
            "question_no": q_no,
            "question": question_text,
            "options": options,
            "answer": answer
        }

        mcqs.append(mcq)

# Print JSON output
print(json.dumps(mcqs, indent=4))
```
